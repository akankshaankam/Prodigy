```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/Users/akankshaankam/Downloads/Prodigy/titanic/data2.csv")
```

```python
data
```

```python
missing_values = data.isnull().sum()
missing_values
```

```python
# Visualize the survival count
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=data)
plt.title('Survival Count')
plt.show()
```

```python
# Visualize the distribution of age
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='Age', bins=20)
plt.title('Distribution of Age')
plt.show()
```

```python
# Relationship between Pclass and Survival
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title('Survival by Pclass')
plt.show()
```

```python
# Relationship between Sex and Survival
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title('Survival by Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
correlation_matrix = data.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Conclusion:
Based on the observations from the visualizations of the Titanic dataset, we can draw the following conclusions:

Survival Count Analysis:
The visualization of survival counts indicates that the number of passengers who did not survive (count labeled as 0) is higher than the number of passengers who did survive (count labeled as 1). This suggests that the majority of passengers in the dataset did not survive the Titanic disaster.
Age Distribution Analysis:
Examining the distribution of ages, we observe that the age range between 20 and 30 years has a notably higher count of passengers. This suggests that a significant portion of the passengers falls within this age group, possibly indicating the demographic composition of the passengers on board.
Pclass and Survival Relationship:
Analyzing the relationship between ticket class (Pclass) and survival, it becomes apparent that the count of passengers in the third class (Pclass 3) is the highest among the ticket classes. This aligns with the observation from the survival count analysis, where the majority of passengers did not survive. The correlation between lower ticket classes and a higher count of passengers who did not survive might indicate a potential link between socio-economic status and survival rate.
Sex and Survival Relationship:
Looking at the relationship between sex and survival, we find that the count of male passengers who did not survive is considerably higher than the count of male passengers who did survive. On the other hand, among female passengers, the count of survivors is higher compared to the count of non-survivors. This stark contrast in survival counts between genders suggests that there might have been prioritization in rescue efforts or other factors that led to differential survival rates between males and females.
In summary, the exploratory data analysis of the Titanic dataset has unveiled several interesting patterns and trends. The dataset indicates that a larger number of passengers did not survive the disaster, with passengers in the third class having the highest count. The age distribution is skewed towards passengers aged 20-30, and there is a notable difference in survival rates based on gender. These findings lay the foundation for more in-depth analyses and provide valuable insights into the dynamics of the Titanic tragedy.
